```python
t = np.linspace(0,50,100)
plt.figure(figsize = (7,4))
plt.plot(t, .5*np.cos(.5*t) + .1 * np.sin(.5 *t), "-", lw = 1, ms = 4, label = "$x(t)$")
plt.plot(t, .5*.1*np.cos(.5*t) - .5*.5 * np.sin(.5*t), "ro-", lw = 1, ms = 4, label = r"$\dot{x(t)}$")
plt.xlabel("$t$", fontsize = 20)
plt.show()
```

```python
A = 10
w = 0.1
B = 0.1
t = np.linspace(0,50,100)
plt.figure(figsize = (7,4))
plt.plot(t, A*np.cos(w*t) + B * np.sin(w *t), "rv", lw = 1, ms = 4, label = "$x(t)$")
plt.plot(t, B*w*np.cos(w*t) - A*w * np.sin(w*t), "bo-", lw = 1, ms = 7, label = r"$\dot{x(t)}$")
plt.xlabel("$t$", fontsize = 20)
plt.show()
```

```python
def theta_t(a, b, g, l, t):
    omega_0 = np.sqrt(g/l)
    return a * np.cos(omega_0 * t) + b * np.sin(omega_0 * t)
```

```python
def pendulo_simple(t = 0):
    fig = plt.figure(figsize = (5,5))
    ax = fig.add_subplot(1, 1, 1)
    x = 2 * np.sin(theta_t(.4, .6, 9.8, 2, t))
    y =  - 2 * np.cos(theta_t(.4, .6, 9.8, 2, t))
    ax.plot(x, y, 'ko', ms = 10)
    ax.plot([0], [0], 'rD')
    ax.plot([0, x ], [0, y], 'k-', lw = 1)
    ax.set_xlim(xmin = -2.2, xmax = 2.2)
    ax.set_ylim(ymin = -2.2, ymax = .2)
    fig.canvas.draw()
interact_manual(pendulo_simple, t = (0, 10,.01));
```
